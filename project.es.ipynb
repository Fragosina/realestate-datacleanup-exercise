#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
# TODO
filas_val = ds[ds['level5'] == "Valdemorillo"]
promedio_val=filas_val['price'].mean()
promedio_val=promedio_val.round(2)

filas_gal= ds[ds['level5'] == "Galapagar"]
promedio_gal=filas_gal['price'].mean()
promedio_gal=promedio_gal.round(2)
print(f"La media de precios en la población de Valdemorillo es de {promedio_val} USD")
print(f"La media de precios en la población de Galapagar es de {promedio_gal} USD")

if promedio_val > promedio_gal:
    resta=promedio_val-promedio_gal
    resta=resta.round(2)
    print(f"Valdemorillo lo supera por {resta} USD")
elif promedio_gal > promedio_val:
    resta=promedio_gal-promedio_val
    resta=resta.round(2)
    print(f"Galapagar lo supera por {resta} USD")
else:
    print("Ambos son iguales")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
# TODO
cinturon_sur['precio_norm'] = (
    cinturon_sur.groupby('level5')['price']
      .transform(lambda x: (x - x.mean()) / x.std())
)

#poblaciones = cinturon_sur['level5'].unique()

poblaciones = cinturon_sur['level5'].unique()[:4]
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

for i, p in enumerate(poblaciones, 1):
    datos = cinturon_sur[cinturon_sur['level5'] == p]['precio_norm'].dropna()
    plt.subplot(2, 2, i)
    plt.hist(datos, bins=30)
    plt.title(p)
    plt.xlabel('Precio normalizado')
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

**TODO: Markdown**. Al ver los histogramas con datos normalizados, podemos apreciar que para casi todas las poblaciones los precios tienden a ser más baratos. Getafe es la que no cumple con eses comportamiento ya que parece haber dos grupos predominantes: uno de casas económicas y otro de casas más caras.

#### Ejercicio 21. Realiza un trazado de las coordenadas (columnas latitud y longitud) del cinturón sur de Madrid por color de cada población (debes usar el subconjunto obtenido del Ejercicio 14) (★★★★)

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps

# Mapa centrado en Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
from ipyleaflet import Map, basemaps, CircleMarker

cinturon_sur['latitude'] = (
    cinturon_sur['latitude']
    .astype(str)
    .str.replace(',', '.', regex=False)
    .astype(float)
)

cinturon_sur['longitude'] = (
    cinturon_sur['longitude']
    .astype(str)
    .str.replace(',', '.', regex=False)
    .astype(float)
)

dicc_coords = (
    cinturon_sur
    .dropna(subset=['level5', 'latitude', 'longitude'])
    .groupby('level5')[['latitude', 'longitude']]
    .apply(lambda x: list(zip(x['latitude'], x['longitude'])))
    .to_dict()
)

colores = {
    'Alcorcón': 'purple',
    'Getafe': 'blue',
    'Fuenlabrada': 'magenta',
    'Leganés': 'cyan'
}

for poblacion, color in colores.items():
    
    # seguridad por si alguna población no existe
    if poblacion not in dicc_coords:
        continue
    
    for lat, lon in dicc_coords[poblacion]:
        map.add(
            CircleMarker(
                location=(lat, lon),
                radius=3,
                color=color,
                fill_color=color,
                fill_opacity=0.5
            )
        )
```
